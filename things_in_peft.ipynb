## dataclass

```python
from dataclasses import dataclass, field, asdict, fields

@dataclass
class PeopleConfig:
    age: int,
    gender: int = field(default=0, metadata={"help": "people's gender"})
    
    @property
    def __dict__(self):
        return asdict(self)
    
    def to_dict(self):
        return self.__dict__
```

```python
pc = PeopleConfig(0)
```

```python
pc.gender
```

```python
fields(pc)[0].metadata
```

```python
import torch.nn as nn

class A(nn.Linear):
    def __init__(self):
        nn.Linear.__init__(self, 10, 20)
        self.weight.requires_grad = False
        print(self.weight.data)
        nn.Linear.reset_parameters(self)
        print(self.weight)
```
